```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/raw/train.csv")  # type: ignore[call-overload]
df.head()
```

```python
# Quick overview of the dataset
df.info()
```

```python
df.describe()
```

```python
# Simple, easy-to-read visualizations
plt.style.use("seaborn-v0_8-whitegrid")

# Survival count
plt.figure(figsize=(6, 4))
df["Survived"].map({0: "Did not survive", 1: "Survived"}).value_counts().reindex([
	"Did not survive",
	"Survived",
]).plot(kind="bar", color=["#d95f02", "#1b9e77"])
plt.title("Survival Count")
plt.xlabel("")
plt.ylabel("Passengers")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

# Age distribution
plt.figure(figsize=(6, 4))
df["Age"].dropna().plot(kind="hist", bins=20, color="#4c78a8", edgecolor="white")
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Passengers")
plt.tight_layout()
plt.show()

# Survival by sex
plt.figure(figsize=(6, 4))
survival_by_sex = df.groupby("Sex")["Survived"].mean().sort_values(ascending=False)
survival_by_sex.plot(kind="bar", color="#66c2a5")
plt.title("Survival Rate by Sex")
plt.xlabel("")
plt.ylabel("Survival rate")
plt.ylim(0, 1)
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
